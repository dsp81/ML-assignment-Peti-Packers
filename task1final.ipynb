Q2)  Do you think we need a machine learning model to differentiate between static activities (laying, sitting, standing) and dynamic activities(walking, walking_downstairs, walking_upstairs)? Look at the linear acceleration (accx^2 + accy^2 + accz^2) for each activity and justify your answer. 

```python
import numpy as np
import matplotlib.pyplot as plt



# Function to load data
def load_data(file_path):
    return np.loadtxt(file_path, delimiter=',', skiprows=1, max_rows=500)

# Function to calculate linear acceleration (without square root)
def calculate_linear_acc(accx, accy, accz):
    return accx**2 + accy**2 + accz**2

# Activity file paths
activities = {
    "Laying": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\LAYING\Subject_2.csv',
    "Sitting": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\SITTING\Subject_2.csv',
    "Standing": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\STANDING\Subject_2.csv',
    "Walking": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING\Subject_2.csv',
    "Walking Downstairs": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING_DOWNSTAIRS\Subject_2.csv',
    "Walking Upstairs": r'c:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING_UPSTAIRS\Subject_2.csv'
}

# Plot setup
fig, axs = plt.subplots(1, 6, figsize=(24, 6))

# To store the average values
average_values = {}

# Loop through activities and plot their linear acceleration
for idx, (activity, path) in enumerate(activities.items()):
    # Load data
    data = load_data(path)
    accx = data[:, 0]
    accy = data[:, 1]
    accz = data[:, 2]
    
    # Calculate linear acceleration (without square root)
    linear_acc = calculate_linear_acc(accx, accy, accz)
    
    # Plot linear acceleration for the activity
    axs[idx].plot(linear_acc, label='Linear Acceleration', color='b')
    axs[idx].set_title(activity)
    axs[idx].set_ylabel('Linear Acceleration')
    axs[idx].legend()
    
    # Calculate and store the average value
    avg_value = np.mean(linear_acc)
    average_values[activity] = avg_value


plt.tight_layout()
plt.show()

print("\nAverage Linear Acceleration for Each Activity:")
for activity, avg_value in average_values.items():
    print(f"{activity}: {avg_value:.2f}")
```


Static Activities:

Linear acceleration is low and stable, averaging between 1.01 and 1.08.

Dynamic Activities:

Linear acceleration is higher and more variable, with walking downstairs showing the most variation, averaging around 1.20.
Observation: While static and dynamic activities have distinct acceleration patterns, some overlap exists. If individual axes were considered separately, a machine learning model might not be necessary for classification.

3.1) Use PCA (Principal Component Analysis) on Total Acceleration (accx^2 + accy^2 + accz^2) to compress the acceleration timeseries into two features and plot a scatter plot to visualize different class of activities.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import pandas as pd
import seaborn as sns

# File paths
file_paths = {
    "laying": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\LAYING\Subject_4.csv',
    "sitting": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\SITTING\Subject_2.csv',
    "standing": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\STANDING\Subject_9.csv',
    "walking": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING\Subject_12.csv',
    "walking_down": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING_DOWNSTAIRS\Subject_24.csv',
    "walking_up": r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\Combined\Test\WALKING_UPSTAIRS\Subject_18.csv'
}

# Load the datasets
dataframes = {key: pd.read_csv(path) for key, path in file_paths.items()}

# Calculate total acceleration magnitude
def compute_total_accel(data):
    data['total_accel'] = np.sqrt(data['accx']**2 + data['accy']**2 + data['accz']**2)
    return data

# Apply the function to all dataframes
dataframes = {key: compute_total_accel(df) for key, df in dataframes.items()}

# Assign labels for each activity
labels = {'laying': 6, 'sitting': 4, 'standing': 5, 'walking': 1, 'walking_down': 3, 'walking_up': 2}
for activity, label in labels.items():
    dataframes[activity]['Activity_Label'] = label

# Combine all datasets
combined_data = pd.concat(dataframes.values(), ignore_index=True)

# Prepare data for PCA
pca_data = pd.DataFrame()
for label in labels.values():
    segment = combined_data[combined_data['Activity_Label'] == label]['total_accel'].values
    segment = segment[:len(segment) - (len(segment) % 500)].reshape(-1, 500)
    pca_model = PCA(n_components=2)
    pca_results = pca_model.fit_transform(segment)
    temp_df = pd.DataFrame(pca_results, columns=['PC1', 'PC2'])
    temp_df['Activity_Label'] = label
    pca_data = pd.concat([pca_data, temp_df], ignore_index=True)

# Map labels to activity names
activity_mapping = {1: 'WALKING', 2: 'WALKING_UPSTAIRS', 3: 'WALKING_DOWNSTAIRS', 4: 'SITTING', 5: 'STANDING', 6: 'LAYING'}
pca_data['Activity_Name'] = pca_data['Activity_Label'].map(activity_mapping)

# Plot the results
sns.set(style='whitegrid')
plt.figure(figsize=(19, 9))
sns.scatterplot(x='PC1', y='PC2', hue='Activity_Name', data=pca_data, s=300, palette='tab10')
plt.title('PCA Scatter Plot for Total Acceleration', fontweight='bold', fontsize=20)
plt.xlabel('Principal Component 1', fontweight='bold', fontsize=20)
plt.ylabel('Principal Component 2', fontweight='bold', fontsize=20)
plt.legend(title='Activities')
plt.show()
```

3.2)  Next, use TSFEL (a featurizer library) to create features (your choice which ones you feel are useful) and then perform PCA to obtain two features. Plot a scatter plot to visualize different class of activities.

```python
import tsfel
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt

# Extract features using TSFEL with domain configuration
feature_config = tsfel.get_features_by_domain()
features_extracted = tsfel.time_series_features_extractor(feature_config, combined_df[['accx', 'accy', 'accz']], window_size=500, verbose=0)

# Align labels with the extracted features
window_count = features_extracted.shape[0]
features_extracted["Activity_Label"] = combined_df['Label'].iloc[::500].head(window_count).values

# Perform PCA on the extracted features
pca_model = PCA(n_components=2)
pca_result = pca_model.fit_transform(features_extracted.drop(columns=["Activity_Label"]))

# Create a DataFrame with PCA results
pca_df = pd.DataFrame(pca_result, columns=["Component_1", "Component_2"])
pca_df["Activity_Label"] = features_extracted["Activity_Label"]

# Map numeric labels to activity names
activity_names = {1: "WALKING", 2: "WALKING_UPSTAIRS", 3: "WALKING_DOWNSTAIRS", 4: "SITTING", 5: "STANDING", 6: "LAYING"}
pca_df["Activity"] = pca_df["Activity_Label"].map(activity_names)

# Plotting the results using Seaborn
sns.set(style="white")
plt.figure(figsize=(18, 8))
sns.scatterplot(x="Component_1", y="Component_2", hue="Activity", data=pca_df, s=250, palette="deep")
plt.title("PCA Visualization of TSFEL Features", fontweight='bold', fontsize=18)
plt.xlabel("Principal Component 1", fontweight='bold', fontsize=16)
plt.ylabel("Principal Component 2", fontweight='bold', fontsize=16)
plt.legend(title="Activity Types")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import seaborn as sns

# File paths for the dataset
acc_x_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_acc_x_train.txt'
acc_y_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_acc_y_train.txt'
acc_z_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_acc_z_train.txt'

gyro_x_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_gyro_x_train.txt'
gyro_y_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_gyro_y_train.txt'
gyro_z_train_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\Inertial Signals\body_gyro_z_train.txt'

labels_path = r'C:\Users\DELL\OneDrive\Desktop\ml_1\es335-24-fall-assignment-1-master\es335-24-fall-assignment-1-master\HAR\UCI HAR Dataset\train\y_train.txt'
labels = pd.read_csv(labels_path, delim_whitespace=True, header=None)

# Function to load and preprocess the data
def read_file(path):
    return pd.read_csv(path, delim_whitespace=True, header=None)

# Load datasets
acc_x = read_file(acc_x_train_path)
acc_y = read_file(acc_y_train_path)
acc_z = read_file(acc_z_train_path)

gyro_x = read_file(gyro_x_train_path)
gyro_y = read_file(gyro_y_train_path)
gyro_z = read_file(gyro_z_train_path)

# Combine the datasets into a single DataFrame
all_features = pd.concat([acc_x, acc_y, acc_z, gyro_x, gyro_y, gyro_z], axis=1)

# Perform PCA
pca_model = PCA(n_components=2)
pca_components = pca_model.fit_transform(all_features)
pca_dataframe = pd.DataFrame(pca_components, columns=["PC1", "PC2"])

# Add labels to the PCA result
pca_dataframe["Activity_Label"] = labels.values.flatten()

# Mapping of numeric labels to activity names
activity_mapping = {1: "WALKING", 2: "WALKING_UPSTAIRS", 3: "WALKING_DOWNSTAIRS", 4: "SITTING", 5: "STANDING", 6: "LAYING"}
pca_dataframe["Activity"] = pca_dataframe["Activity_Label"].map(activity_mapping)

# Plotting the PCA result
sns.set(style="ticks")
plt.figure(figsize=(18, 8))
sns.scatterplot(x="PC1", y="PC2", hue="Activity", data=pca_dataframe, s=250, palette="Set2")
plt.title("Activity Classification with PCA", fontsize=18, fontweight='bold')
plt.xlabel("Principal Component 1", fontsize=16, fontweight='bold')
plt.ylabel("Principal Component 2", fontsize=16, fontweight='bold')
plt.legend(title="Activity Types")
plt.show()
```
